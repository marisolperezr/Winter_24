```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('alzheimers_disease_data.csv')
```

```python
#relationship between age and alzheimer's diagnose
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Age', hue='Diagnosis', data=df, palette="pastel", width=0.7)
plt.title('Relationship between Age and Diagnose of Alzheimer', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#The most frequent age 
```

```python
# We have to filter the diagnose because we only want patients whith alzheimer (Diagnosis == 1)
alzheimers_patients = df[df['Diagnosis'] == 1]

age_alzheimer_count = alzheimers_patients[['Age', 'Diagnosis']].groupby('Age').size().reset_index(name='Count')
top_age = age_alzheimer_count.sort_values(by='Count', ascending=False).head(1)
print("The age with more alzheimer diagnoses is :")
print(top_age)

```

```python
# Seeing if how many alzheimer's patients do smoke or not
```

```python
alzheimers_patients = df[df['Diagnosis'] == 1]
#counting how many patients
smoking_alzheimer_count = alzheimers_patients['Smoking'].value_counts()

plt.figure(figsize=(8,6))
sns.barplot(x=smoking_alzheimer_count.index, y=smoking_alzheimer_count.values)
plt.title('Smokers and Non-smokers Alzheimer patients', fontsize=14,fontweight='bold')
plt.xlabel('Smoking Habit?', fontsize=12,fontweight='bold')
plt.ylabel('Number of alzheimer patients', fontsize=12,fontweight='bold')
plt.xticks([1, 0], ['Non-Smokers', 'Smokers']) 
plt.show()
```

```python
#This graphic shows hoy many patients of alzheimer have each educarion level
```

```python
count_el_by_diagnose = df.groupby('EducationLevel')['Diagnosis'].sum()
count_el_by_diagnose.sort_values(ascending=False).plot(kind='bar',color='lightblue')
plt.title('Education Level vs Alzheimers Diagnose', fontweight='bold')
plt.xlabel('Education Level', fontweight='bold')
plt.ylabel('Patients', fontweight='bold')
plt.xticks([0, 1, 2, 3], ['None', 'Secondary', 'Licenciature', 'Superior'])
plt.show()
```

```python
#Relationship between physical activity with having alzheimer
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Diagnosis', y='PhysicalActivity', data=df, color="pink")
plt.title('Relationship between Physical activity and Alzheimer Diagnose', fontsize=14, fontweight='bold')
plt.xlabel('Alzheimer Diagnose', fontsize=12, fontweight='bold')
plt.ylabel('Physical activity (hours per week)', fontsize=12,fontweight='bold')
plt.show()
```

```python
#Relationship between having alzheimer and having a family member with alzheimer
```

```python
alzheimers_patients = df[df['Diagnosis'] == 1]
alzheimers_with_history = alzheimers_patients[alzheimers_patients['FamilyHistoryAlzheimers'] == 1].shape[0]
alzheimers_without_history = alzheimers_patients[alzheimers_patients['FamilyHistoryAlzheimers'] == 0].shape[0]

history_comparison = {'No History': alzheimers_with_history, 'History': alzheimers_without_history}
plt.figure(figsize=(8,6))
sns.barplot(x=list(history_comparison.keys()), y=list(history_comparison.values()), color="pink")

plt.title('Alzheimer & Family History', fontsize=14, fontweight='bold')
plt.xlabel('Family history', fontsize=12, fontweight='bold')
plt.ylabel('Number of Alzheimer Patients', fontsize=12, fontweight='bold')
plt.xticks(rotation=25)
plt.show()
```

```python
#Comparison between alzheimers patients who have or not diabetes
```

```python
alzheimers_patients = df[df['Diagnosis'] == 1]
alzheimers_with_diabetes = alzheimers_patients[alzheimers_patients['Diabetes'] == 1].shape[0]
alzheimers_without_diabetes = alzheimers_patients[alzheimers_patients['Diabetes'] == 0].shape[0]

diabetes_comparison = {'No Diabetes': alzheimers_with_diabetes, 'Diabetes': alzheimers_without_diabetes}
plt.figure(figsize=(8,6))
sns.barplot(x=list(diabetes_comparison.keys()), y=list(diabetes_comparison.values()), color="lightgreen")

plt.title('Alzheimer & Diabetes', fontsize=14, fontweight='bold')
plt.xlabel('Diabetes', fontsize=12, fontweight='bold')
plt.ylabel('Number of Alzheimer Patients', fontsize=12, fontweight='bold')
plt.xticks(rotation=25)
plt.show()
```
