```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
jobs=pd.read_csv('jobs.csv')
```

```python
jobs.job_title.value_counts()
```

```python
jobs.job_title.value_counts().plot(kind="bar")
```

```python
jobs.job_title.unique()
```

```python
df=jobs.job_title.head(20)
```

```python
df.value_counts().plot(kind="bar", color="orange")
```

```python
#Top 5 more frequent job titles
top_jobs = jobs.job_title
```

```python
top_jobs.value_counts().head(5).plot(kind="bar", color="pink")
plt.title("TOP 5 JOB TITLES", fontweight="bold")
plt.xlabel("Job title",fontweight="bold")
plt.ylabel("Frequency",fontweight="bold")
plt.show()
```

```python
#Top 10 highest paid jobs
```

```python
top10 =jobs[['job_title', 'salary_in_usd']].sort_values(by='salary_in_usd',ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10["job_title"], top10["salary_in_usd"], color="pink")
plt.title("Top 10 highest-paid jobs", fontweight="bold")
plt.xlabel("Salary_in_usd",fontweight="bold")
plt.ylabel("Job Title",fontweight="bold")
plt.show()
```

```python
#Highest paid job
```

```python
top_paid_job = jobs[['job_title', 'salary_in_usd']].sort_values(by='salary_in_usd', ascending=False).head(1)
print("The highest paid job is:")
print(top_paid_job)
```

```python
#Top 10 worst paid jobs
```

```python
top10worstpaid =jobs[['job_title', 'salary_in_usd']].sort_values(by='salary_in_usd',ascending=True).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10worstpaid["job_title"],top10worstpaid["salary_in_usd"], color="pink")
plt.title("Top 10 worst paid jobs", fontweight="bold")
plt.xlabel("Salary_in_usd", fontweight="bold")
plt.ylabel("Job Title", fontweight="bold")
plt.show()
```

```python
#Worst paid job
```

```python
worst_paid_job = jobs[['job_title', 'salary_in_usd']].sort_values(by='salary_in_usd', ascending=True).head(1)
print("The worst paid job is:")
print(worst_paid_job)
```

```python
#Wich job Category has the highest salaries
```

```python
avg_salary_by_category = jobs.groupby('job_category')['salary_in_usd'].mean()
avg_salary_by_category.sort_values(ascending=False).plot(kind='bar', figsize=(10,6), color="lightgreen")
plt.title('Average Salary by Job Category', fontweight='bold')
plt.xlabel('Job Category', fontweight='bold')
plt.ylabel('Average Salary (USD)', fontweight='bold')
plt.show()
```

```python
#How salaries change according to the experience level. Average salary by experience level
```

```python
average_salary_by_experience = jobs.groupby('experience_level')['salary_in_usd'].mean()
average_salary_by_experience.sort_values(ascending=False).plot(kind='bar',color='lightblue')
plt.title('Average Salary by Experience Level', fontweight='bold')
plt.xlabel('Experience Level', fontweight='bold')
plt.ylabel('Average Salary', fontweight='bold')
plt.show()
```

```python
#Salaries depending on the location
```

```python
average_salary_by_location = jobs.groupby('company_location')['salary_in_usd'].mean()
average_salary_by_location.sort_values(ascending=False)
average_salary_by_location.plot(kind='bar', figsize=(12,8), color='pink')
plt.title("Average Salary by Company Location", fontweight='bold')
plt.xlabel("Company Location", fontweight='bold')
plt.ylabel("Average Salary", fontweight='bold')
plt.show()
```

```python
#Top 10 highest paid countries
```

```python
average_salary_by_residence = jobs.groupby('employee_residence')['salary_in_usd'].mean()
average_salary_by_residence.sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,6), color='turquoise')
plt.title('Average Salary by Employee Residence', fontweight='bold')
plt.xlabel('Employee Residence', fontweight='bold')
plt.ylabel('Average Salary (USD)', fontweight='bold')
plt.show()
```

```python
#top 10 worst paid countries
```

```python
average_salary_by_residence = jobs.groupby('employee_residence')['salary_in_usd'].mean()
average_salary_by_residence.sort_values(ascending=True).head(10).plot(kind='bar', figsize=(10,6), color='lightgreen')
plt.title('Average Salary by Employee Residence', fontweight='bold')
plt.xlabel('Employee Residence', fontweight='bold')
plt.ylabel('Average Salary (USD)', fontweight='bold')
plt.show()
```

```python
#Relationship between salary and experience level
```

```python
average_salary_by_experience = jobs.groupby('experience_level')['salary_in_usd'].mean()

plt.figure(figsize=(10, 6))
average_salary_by_experience.sort_values(ascending=True).plot(kind='bar', color='skyblue')
plt.title('Average Salary by Experience Level', fontweight='bold')
plt.xlabel('Experience Level', fontweight='bold')
plt.ylabel('Average Salary', fontweight='bold')
plt.xticks(rotation=45)

plt.show()
```
